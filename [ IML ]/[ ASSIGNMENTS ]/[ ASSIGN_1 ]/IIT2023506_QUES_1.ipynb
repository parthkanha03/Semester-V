# Task
Deduce the Normal Equation for a linear regression model with two predictors, then use it to fit a least squares line to the provided annual revenue data, predict the revenue for 2025, and analyze the expected error in predictions.

Annual Revenue data:
Y | 2014 | 2015 | 2016 | 2017 | 2018 | 2019 | 2020 | 2021 | 2022 | 2023 | 2024  
Rev. In billion Rupees | 200.5 | 198.5 | 187.4 | 189.2 | 188.9 | 183.5 | 189.1 | 284 | 292.3 | 396 | 397

```python
import pandas as pd

data = {
    'Year': [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    'Revenue': [200.5, 198.5, 187.4, 189.2, 188.9, 183.5, 189.1, 284.0, 292.3, 396.0, 397.0]
}

df = pd.DataFrame(data)
display(df)
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.vstack([np.ones(len(df)), df['Year']]).T
y = df['Revenue'].values

beta_hat = np.linalg.inv(X.T @ X) @ X.T @ y

years_plot = np.arange(df['Year'].min(), df['Year'].max() + 2, 1)
X_plot = np.vstack([np.ones(len(years_plot)), years_plot]).T
y_pred = X_plot @ beta_hat

plt.figure(figsize=(10, 6))
plt.scatter(df['Year'], df['Revenue'], label='Original Data')
plt.plot(years_plot, y_pred, color='red', label='Fitted Line')
plt.xlabel('Year')
plt.ylabel('Revenue (in billion Rupees)')
plt.title('Annual Revenue with Fitted Least Squares Line')
plt.legend()
plt.show()
```

```python

year_2025 = np.array([1, 2025])

predicted_revenue_2025 = year_2025 @ beta_hat

print(f"Predicted Revenue for 2025: {predicted_revenue_2025:.2f} billion Rupees")
```
